# Variables, Expressions & Statements
-  For now think of them as you would algebraic variables, late we'll go deeper into objects
-  Do keep in mind that many variables point to objects that are immutable (can't be changed in place)
-  Expressions are code that evaluate to some value
-  statements combine, often assigning, expression results (values) to variables
-  variables (objects) have a specific type, i.e. number or string, chosen dynamically at assignment
-  variables (and all names in Python) are case-sensitive

```python
# x is the name of an object (variable)
# 2 is a simple, literal expression
# = is the assignment operator; 2 is assigned to the object (variable) named x
x = 2
y = 2
z = x + y
print(z)
# type is a built-in function that tells us what type (integer, string, etc) of object we have
print(type(z))
# every object in memory has a unique id, the built-in function id tells us what it is
print(id(z))

#assign new value to x, creates a new object
x = 2.0
z = x + y
print(z)
print(type(z))
print(id(z))
```

## Expressions & Numeric Operators

```python
# assign 2 integers to two objects and do arithmetic
x = 2
y = 5
z = y-x
print(z)
#std div
z=x/y
print(z)
z = y/x
print(z)
#floor (int) div
z = y//x
print(z)
#modulus
z = y%2
print(z)
#exp
z=x**y
print(z)
```

## strings
-  contains text, enclosed in single, double or triple quotes
-  numeric values in quotes are strings, but can be treated as numbers
-  keep in mind when using the input function
-  This is just a brief intro, later we'll see there are many string methods (functions) available

```python
#strings
# operator overload
x = '2'
y = '5'
z = x + y
print(z)

message = 'Hello World'
print(message)

messages = message * 2
print(messages)

#sring method demo
print(message.upper())

#printing
print("The answer is",z)
```

## 'convert' (aka cast) string to numeric or vice versa

```python
x = '2'
y = 4
# print(x + y)  # uncomment to get an error
print(int(x) + y)
print(float(x))

# This one is tricky, x is a string, and were converting y to a string
# result is a string of 2 & 4 concatenated together into the string '24'
print(x + str(y))
```

## prompting for input
-  useful for short input
-  remember all input from keyboard is string,
-  needs to be converetd to number if numeric input is needed

```python
#name = input("What is your name? ")
#print(name)

x = input("value 1: ")
y = input("value 2: ")
print(x + y)

#using int / float
x = int(input("value 1: "))
y = float(input("value 2: "))
print(x + y)
```
